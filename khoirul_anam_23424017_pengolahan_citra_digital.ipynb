```python
# ==========================================
# PRAKTIKUM PENGOLAHAN CITRA DIGITAL
# RGB -> GRAYSCALE -> BINARY -> MBIT
# BRIGHTNESS -> CONTRAST -> HISTOGRAM -> EQUALIZATION
# ==========================================

import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# ------------------------------------------
# Upload Gambar
# ------------------------------------------
print("Silakan upload gambar")
uploaded = files.upload()

# membaca gambar
img = cv2.imread(list(uploaded.keys())[0])
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# ------------------------------------------
# 1. RGB ke Grayscale
# ------------------------------------------
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# ------------------------------------------
# 2. Grayscale ke Binary
# ------------------------------------------
ret, binary = cv2.threshold(gray,127,255,cv2.THRESH_BINARY)

# ------------------------------------------
# 3. Grayscale ke m-bit
# ------------------------------------------
m = 4
levels = 2**m
mbit = np.floor(gray/(256/levels))*(256/levels)

# ------------------------------------------
# 4. Brightness
# ------------------------------------------
brightness = cv2.convertScaleAbs(gray, alpha=1, beta=50)

# ------------------------------------------
# 5. Contrast
# ------------------------------------------
contrast = cv2.convertScaleAbs(gray, alpha=1.5, beta=0)

# ------------------------------------------
# 6. Histogram Equalization
# ------------------------------------------
equalized = cv2.equalizeHist(gray)

# ------------------------------------------
# MENAMPILKAN HASIL
# ------------------------------------------

plt.figure(figsize=(12,10))

plt.subplot(3,3,1)
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis("off")

plt.subplot(3,3,2)
plt.imshow(gray, cmap='gray')
plt.title("Grayscale")
plt.axis("off")

plt.subplot(3,3,3)
plt.imshow(binary, cmap='gray')
plt.title("Binary")
plt.axis("off")

plt.subplot(3,3,4)
plt.imshow(mbit, cmap='gray')
plt.title("Grayscale m-bit")
plt.axis("off")

plt.subplot(3,3,5)
plt.imshow(brightness, cmap='gray')
plt.title("Brightness")
plt.axis("off")

plt.subplot(3,3,6)
plt.imshow(contrast, cmap='gray')
plt.title("Contrast")
plt.axis("off")

plt.subplot(3,3,7)
plt.hist(gray.ravel(),256,[0,256])
plt.title("Histogram")

plt.subplot(3,3,8)
plt.imshow(equalized, cmap='gray')
plt.title("Histogram Equalization")
plt.axis("off")

plt.tight_layout()
plt.show()
```
